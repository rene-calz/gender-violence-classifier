```python
import pandas as pd
import numpy as np

from sklearn.compose import make_column_transformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('/content/df_sin_lugar.csv')
df.head()
```

```python
df.info()
```

```python
X = df.drop(columns=['VFIS_12M_2','Unnamed: 0'])
y = df['VFIS_12M_2']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 1234)
```

##LogisticRegressionClassifier

```python
lr = LogisticRegression(penalty = None, solver = "saga", max_iter = 10000)
```

```python
lr_fitted = lr.fit(X_train, y_train)
```

```python
lr_fitted.score(X_test, y_test)
```

```python
column_names = list(df.drop(columns=['VFIS_12M_2','Unnamed: 0']).columns.values)
column_names
```

```python
print(len(column_names))
```

```python
coef_abs = [x for x in lr_fitted.coef_[0]]
```

```python
coef = [(x,y) for x,y in enumerate(lr_fitted.coef_[0])]
```

```python
print(len(coef))
```

```python
coeficients = pd.DataFrame({'Variables': column_names, 'Coef': coef_abs})
coeficients.head()
```

```python
coeficients.sort_values(by='Coef', axis=0, ascending=False).head()
```

##4-fold KNN

```python
knn_pipeline = make_pipeline(knn)
```

```python
training = pd.concat([X_train, y_train], axis = 1)
```

```python
subtraining, validation = train_test_split(training, test_size = 0.25, stratify = training['VFIS_12M_2'], random_state = 1234)
```

```python
X_sub = subtraining.drop(columns = ['VFIS_12M_2'])
y_sub = subtraining["VFIS_12M_2"]
```

```python
knn_fitted = knn_pipeline.fit(X_sub,y_sub)
```

```python
knn_pipeline.score(validation.drop(columns="VFIS_12M_2"), validation["VFIS_12M_2"])
```

```python
from sklearn.model_selection import cross_validate

cv_4 = cross_validate(estimator = knn_pipeline, #Modelo o Pipeline que usaremos
                      cv = 4, #NÃºmero de folds
                      X = X_train,
                      y = y_train)
```

```python
res = pd.DataFrame(cv_4)
res.head()
```

```python
res.agg(["mean", "sem"])
```

```python
knn = KNeighborsClassifier()
```

```python
knn_tune_pipe = make_pipeline(knn)
```

```python
knn_tune_pipe.get_params()
```

```python
parameter_grid = {"kneighborsclassifier__n_neighbors": range(1, 50)}
```

```python
from sklearn.model_selection import GridSearchCV

knn_tune_grid = GridSearchCV(estimator = knn_tune_pipe,
                             param_grid = parameter_grid,
                             cv = 10)
```

```python
knn_tuned = knn_tune_grid.fit(X_train, y_train)
```

```python
knn_tune_grid.best_params_
```

```python
accuracies_grid = pd.DataFrame(knn_tune_grid.cv_results_)
accuracies_grid
```

```python
accuracies_grid["sem_test_score"] = accuracies_grid["std_test_score"] / 10**(1/2)
```

```python
accuracies_grid = accuracies_grid[["param_kneighborsclassifier__n_neighbors",
                                   "mean_test_score",
                                   "sem_test_score"]]
```

```python
accuracies_grid.rename(columns = {"param_kneighborsclassifier__n_neighbors": "n_neighbors"}, inplace = True)
accuracies_grid
```

```python
accuracies_grid.nlargest(1, "mean_test_score")
```

```python
y_pred = knn_tune_grid.predict(X_test)
```

```python
knn_tune_grid.score(X_test,y_test)
```
